# Outliers

```python
# Define the function to remove outliers
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# List of columns to clean
columns_to_clean = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']


for column in columns_to_clean:
    data = remove_outliers(data, column)

# Display the first few rows of the cleaned dataset
print(data.head())
```

**One-Hot Encoding**

```python
one_hot_columns=['income','sex']
data_encoded=pd.get_dummies(data, columns=one_hot_columns)
```

**Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_columns = ['workclass', 'education', 'marital_status', 'occupation', 'relationship', 'race', 'native_country']
label_encoders = {}
for col in label_columns:
    label_encoders[col] = LabelEncoder()
    data_encoded[col] = label_encoders[col].fit_transform(data[col])

```

```python
print(data_encoded.head())
```

```python
data_encoded['capital_dff'] = data_encoded['capital_gain'] - data_encoded['capital_loss']
data_encoded['work_hours_per_year'] = data_encoded['hours_per_week'] * 52
data_encoded['capital_gain_log'] = np.log1p(data_encoded['capital_gain'])

```

# Isolation Forest

```python
from sklearn.ensemble import IsolationForest
```

```python
numerical_features=['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week', 'work_hours_per_year', 'capital_gain_log']
```

```python
iso_forest=IsolationForest(contamination=0.01, random_state=42)
outliers=iso_forest.fit_predict(data_encoded[numerical_features])
```

```python
data_encoded['outliers'] = outliers
```

```python
data_cleaned = data_encoded[data_encoded['outliers'] == 1].drop(columns=['outliers'])
```
