This Describe

Count: The number of non-missing (non-NaN) values.

Mean: The average value of the column.

Std (Standard Deviation): Measures the spread of the data points around the mean.

Min: The minimum value in the column.

25% (1st Quartile): The value below which 25% of the data points fall.

50% (Median or 2nd Quartile): The middle value of the data, below which 50% of the data points fall.

75% (3rd Quartile): The value below which 75% of the data points fall.

Max: The maximum value in the column.

New we are replacing every negative value with mean with the help of for loop

```python
# Looping through each and every column to replace negative values with the median of non-negative values
for column in data_cleaned.columns:
    if data_cleaned[column].min() < 0:
        median_value = data_cleaned[column][data_cleaned[column] >= 0].median()
        data_cleaned.loc[:, column] = data_cleaned[column].apply(lambda x: x if x >= 0 else median_value)
```

After replcaing negative values with mediane we got

Standard Deviation (std): This will likely decrease, as replacing outliers (negative values) with median values reduces the spread of data.

Minimum (min): The minimum value will increase from a negative value to the smallest positive value or the median used for replacement.

Maximum (max): Generally remains the same unless there were negative outliers that were significant in magnitude.

Quartiles (25%, 50%, 75%): These could shift slightly depending on how the negative values affected the distribution of data.

Calculate Median: median_value stores the median of the 'DS' column.

Replace Negatives: The .apply() function replaces all negative values in the 'DS' column with the median.

```python
median_value = data_cleaned['DS'].median()
data_cleaned.loc[:, 'DS'] = data_cleaned['DS'].apply(lambda x: median_value if x < 0 else x)
print(data_cleaned['DS'])
```

Calculate Z-scores: Computes Z-scores for each value in the 'DS' column.

Identify Outliers: outliers is a boolean array where True indicates values that are more than 3 standard deviations from the mean.

Replace Outliers: Sets outliers to the median value.

```python
data_cleaned = data_cleaned.copy()
z_scores = stats.zscore(data_cleaned['DS'])
outliers = (abs(z_scores) > 3)
data_cleaned.loc[outliers, 'DS'] = median_value
print(data_cleaned['DS'])
```

Calculate Q1 (25th Percentile): Q1 is the value below which 25% of the data falls.

Calculate Q3 (75th Percentile): Q3 is the value below which 75% of the data falls.

Calculate IQR: IQR is the interquartile range, calculated as Q3 - Q1

Lower Bound: Any value below this is considered an outlier.

Upper Bound: Any value above this is considered an outlier

Cap Upper Values: Sets any value greater than upper_bound to upper_bound.

Cap Lower Values: Sets any value less than lower_bound to lower_bound

```python
data_cleaned = data_cleaned.copy()
Q1 = data_cleaned['DS'].quantile(0.25)
Q3 = data_cleaned['DS'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

data_cleaned.loc[data_cleaned['DS'] > upper_bound, 'DS'] = upper_bound
data_cleaned.loc[data_cleaned['DS'] < lower_bound, 'DS'] = lower_bound
print(data_cleaned['DS'])  # Verify capping
```
